```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Charger le dataset

df = pd.read_csv('Speed Dating Data.csv', encoding= 'unicode_escape')
```

```python
# Option pour voir la totalit√© de nos colonnes

pd.set_option('display.max_column', 195)
```

```python
# Visualiser la taille de notre dataset

df.shape
```

```python
# Aper√ßu g√©n√©ral

df.head(10)
```

```python
# Observer la r√©partition de nos features en fonction des types

df.dtypes.value_counts().plot.pie()
```

```python
# Aper√ßu g√©n√©ral des valeurs manquantes via une heatmap

sns.heatmap(df.isnull(), cbar=True)
sns.set(rc={'figure.figsize':(20,10)})
```

```python
# 1 √®re m√©thode : On supprime les features avec un nombre trop √©lev√© de valeurs manquantes en rentrant le nom de la colonne 

df =  df.drop(columns = ['attr1_s','fun1_s','attr3_s','fun3_s'])
```

```python
# 2 nd m√©thode : On fait du slicing colonnes, en l'occurence les 13 derni√®res

df  = df.drop(df.columns[-13:], axis=1, inplace=False)
```

```python
# On regarde de nouveau la heatmap apr√®s une premi√®re couche de nettoyage

sns.heatmap(df.isnull(), cbar=True)
sns.set(rc={'figure.figsize':(20,10)})
```

```python
# On regarde les features qui pr√©sentent un taux trop important de valeurs manquantes 

(df.isna().sum()/df.shape[0]).sort_values(ascending=True)
```

```python
# Vu la taille imposante du dataset on va supprimer toutes les features qui pr√©sentent plus de 10% de valeurs manquantes

df = df[df.columns[df.isna().sum()/df.shape[0] < 0.1]]
df.shape

# Il nous reste donc 87 colonnes apr√®s ce nettoyage
```

```python
# On va essayer de relever quelques corr√©lation malgr√© le nombre de features encore restant

plt.subplots(figsize=(20,20))

corr = df.corr()
sns.heatmap(corr)
```

```python
# On va zoomer sur les qualit√©es et quelles sont les plus susceptibles de concr√©tiser un 'match'

df1 = pd.concat([df['match'],df['attr'],df['sinc'],df['intel'],df['fun'],df['amb'],
                   df['like'],df['prob'],df['met']],axis=1)
plt.subplots(figsize=(15,10))
ax = plt.axes()
ax.set_title("Heatmap n¬∞ 1")
corr = df1.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)

# Fun et like semblent avoir une bonne corr√©lation avec le match
# On voit √©galement une forte corr√©lation entre certaines qualit√©s
```

```python
# On appercoit une deuxi√®me zone fortement corr√©l√© qui correspond aux activit√©s

df2 = pd.concat([df['match'],df['sports'],df['tvsports'],df['exercise'],df['dining'],df['museums'],
                   df['art'],df['hiking'],df['gaming'],df['reading'],df['tv'],df['theater'],
                    df['movies'],df['concerts'],df['music'],df['shopping'],df['yoga']],axis=1)
plt.subplots(figsize=(15,10))
ax = plt.axes()
ax.set_title("Heatmap n¬∞ 2")
corr = df2.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)

# On ne voit pas de corr√©alation flagrante entre les activit√©s et les 'matchs' 
# N√©anmoins on voit clairement des 'pairs' (dining/shopping ou tv/shopping) 
```

```python
# Pr√©paration d'un nouveau dataset sans les features explor√©es

df3 = df.drop(columns = ['match','attr','sinc','intel','fun','amb',
                   'like','prob','met','match','sports','tvsports','exercise','dining','museums',
                   'art','hiking','gaming','reading','tv','theater',
                    'movies','concerts','music','shopping','yoga'])
```

```python
# Nouvelle heatmap 

corr = df3.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
# Nous allons explorer les domaines d'√©tudes des pr√©tendants
# Pour cela nous allons associer chaque nombre au domaine d'√©tude auquel il correspond

df['field_cd'] = df.field_cd.map({1: 'Law', 2: 'Math', 3: 'Soc. Sc.', 4: 'Med. Sc.', 5: 'Eng.', 6: 'Journ.',
                                  7: 'Hist.', 8: 'Econ', 9: 'Educ.',10: 'Nat. Sc.', 11: 'Soc. Wr.', 12: 'Und.',
                                  13: 'Pol. Sc.', 14: 'Film', 15: 'Arts', 16:'Lang.', 17: 'Arch.', 18: 'Oth.'}).fillna(df.field_cd)
```

```python
# Repr√©sentation graphique entre le nombre de pr√©tendants et leur domaine d'√©tude respectif

plt.figure(figsize = (16,6))
ax = sns.countplot(x="field_cd", data=df)
plt.title("Domaine d'√©tude", fontsize=14)

ax.set_xticklabels(ax.get_xticklabels(),rotation=40)

plt.ylim(0, 2000)
plt.xlabel('')
plt.ylabel('Nombre')
```

```python
# Nous pouvons √©galement regarder le rapport en fonction du genre 

# Nous rempla√ßons 0-1 par Femme-Homme pour une meilleure compr√©hension

df['gender'] = df['gender'].astype(str) 
df['gender'] = df['gender'].replace(str(0), 'Femme')
df['gender'] = df['gender'].replace(str(1), 'Homme')

# Repr√©sentation graphique

mf = df[['gender', 'field_cd']].groupby(['field_cd', 'gender']).size().unstack()
ax = mf.plot(kind='bar', figsize=(16,6), stacked=True)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, rotation=45)

ax.set_title("Domaine d'√©tude par genre", fontsize=14)
ax.set_xlabel('')
ax.set_ylabel('Nombre')
```
